# Penjelasan Dataset "Heart Disease"

## Dataset Heart Disease dari UCI Machine Learning Repository berisi data yang dikumpulkan dari empat institusi medis: Cleveland, Hungary, Switzerland, dan VA Long Beach. Dataset ini digunakan untuk memprediksi ada atau tidaknya penyakit jantung berdasarkan berbagai faktor medis. Total terdapat 303 sampel dengan 14 fitur yang relevan, termasuk atribut seperti usia, jenis kelamin, tekanan darah, kolesterol, dan beberapa lainnya yang mengindikasikan hasil uji medis, seperti hasil tes treadmill (thalach), tingkat angina, dan sebagainya.

## 1. import library

```python
#import library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2. Loading the Dataset

```python
# Menggunakan URL untuk memuat dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
columns = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"]

# Membaca data dengan pandas
data = pd.read_csv(url, names=columns)

# Menampilkan 5 baris pertama untuk melihat data
print(data.head())

import numpy as np

# Mengganti nilai '?' dengan NaN
data.replace('?', np.nan, inplace=True)

# Mengonversi kolom menjadi tipe numerik, setelah mengganti '?'
data = data.apply(pd.to_numeric, errors='coerce')

# Menghapus baris yang mengandung NaN
data.dropna(inplace=True)
```

# 3. Basic Summary Statistics

```python
# Deskripsi statistik dari dataset
print(data.describe())
```

# 4. Distribution Frequency of Each Feature

```python
# Membuat histogram untuk distribusi setiap fitur numerik
data.hist(figsize=(12, 12), bins=20, color='blue', edgecolor='black')
plt.show()
```

# 5. Scatter Plot Dua Fitur

```python
# Scatter plot antara usia dan denyut jantung maksimum
plt.scatter(data['age'], data['thalach'], c=data['target'], cmap='viridis')
plt.title('Scatter Plot antara Usia dan Denyut Jantung Maksimum')
plt.xlabel('Usia')
plt.ylabel('Denyut Jantung Maksimum (thalach)')
plt.show()
```

# 6. Korelasi Antar Fitur

```python
# Heatmap korelasi antar fitur
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Korelasi Antar Fitur Dataset Heart Disease')
plt.show()
```

## Tambahkan EDA lain

# 1. Boxplot untuk Memeriksa Outlier

```python
# Menggunakan boxplot untuk mendeteksi outlier dalam fitur penting seperti kolesterol dan tekanan darah.
plt.figure(figsize=(10,6))
sns.boxplot(data['chol'])
plt.title('Distribusi Kolesterol (chol)')
plt.show()
```

# 2. Analisis Principal Component Analysis (PCA)

```python
# Melakukan PCA untuk melihat apakah fitur-fitur dataset dapat direduksi menjadi komponen yang lebih sedikit dan tetap merepresentasikan variasi data.
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standarisasi data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.drop('target', axis=1))

# Melakukan PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

plt.scatter(pca_data[:, 0], pca_data[:, 1], c=data['target'], cmap='plasma')
plt.title('PCA dari Fitur Dataset Heart Disease')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

## Usia dan jenis nyeri dada memiliki korelasi yang cukup signifikan dengan penyakit jantung. Ada hubungan antara tekanan darah tinggi dan denyut jantung maksimum yang dapat menunjukkan risiko penyakit jantung. Analisis PCA menunjukkan bahwa beberapa fitur dataset bisa direduksi menjadi 2 komponen utama, meski perlu verifikasi lebih lanjut dengan model prediktif.
